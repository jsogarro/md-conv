## Mathematical Operations

Let's perform some calculations:

```python
values = [1, 2, 3, 4, 5]
total = sum(values)
average = total / len(values)
print(f'Sum: {total}, Average: {average}')
```

```python
# Final computation
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(f'5! = {factorial(5)}')
```
